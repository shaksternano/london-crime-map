```python
import pandas as pd

crime_data = pd.read_excel("data/london-crime-dataset-01-04-2021-to-31-03-2025.xlsx")
crime_data
```

```python
boroughs = [
    "Barking and Dagenham",
    "Barnet",
    "Bexley",
    "Brent",
    "Bromley",
    "Camden",
    "Croydon",
    "Ealing",
    "Enfield",
    "Greenwich",
    "Hackney",
    "Hammersmith and Fulham",
    "Haringey",
    "Harrow",
    "Havering",
    "Hillingdon",
    "Hounslow",
    "Islington",
    "Kensington and Chelsea",
    "Kingston upon Thames",
    "Lambeth",
    "Lewisham",
    "Merton",
    "Newham",
    "Redbridge",
    "Richmond upon Thames",
    "Southwark",
    "Sutton",
    "Tower Hamlets",
    "Waltham forest",
    "Wandsworth",
    "Westminster",
]

capitalize_exceptions = ["a", "an", "and", "of", "in", "on", "for", "the", "to", "with"]


def get_borough(borough_snt: str) -> str:
    lowered_borough_snt = borough_snt.lower()
    for borough_name in boroughs:
        if borough_name.lower() in lowered_borough_snt:
            return borough_name
    return "Unknown"


def fix_case(string: str) -> str:
    split = string.split(" ")
    for i, word in enumerate(split):
        if word.lower() in capitalize_exceptions:
            split[i] = word.lower()
        else:
            split[i] = word.capitalize()
    return " ".join(split)


crime_data["Borough"] = crime_data["Borough_SNT"].apply(get_borough)

crime_data["Offence Group Fixed"] = crime_data["Offence Group"].apply(fix_case)
crime_data["Offence Subgroup Fixed"] = crime_data["Offence Subgroup"].apply(fix_case)

crime_data
```

```python
crime_by_date = crime_data.groupby("Month_Year")
crime_by_date.size()
```

```python
import json

known_boroughs = crime_data[crime_data["Borough"] != "Unknown"]
processed_data = {
    "total": crime_data["Count"].sum().item(),
    "total_known_borough": known_boroughs["Count"].sum().item(),
    "dates": {},
}

for date, date_group in crime_by_date:
    date_string = date.strftime("%Y-%m-%d")
    known_boroughs_by_date = date_group[date_group["Borough"] != "Unknown"]
    processed_data["dates"][date_string] = {
        "total": date_group["Count"].sum().item(),
        "total_known_borough": known_boroughs_by_date["Count"].sum().item(),
        "boroughs": {},
    }
    for borough, borough_group in date_group.groupby("Borough"):
        processed_data["dates"][date_string]["boroughs"][borough] = {
            "total": borough_group["Count"].sum().item(),
            "offence_groups": {},
        }
        for offence_group, offence_group_group in borough_group.groupby("Offence Group Fixed"):
            processed_data["dates"][date_string]["boroughs"][borough]["offence_groups"][offence_group] = {
                "total": offence_group_group["Count"].sum().item(),
                "offence_subgroups": {},
            }
            for offence_subgroup, offence_subgroup_group in offence_group_group.groupby("Offence Subgroup Fixed"):
                processed_data["dates"][date_string]["boroughs"][borough]["offence_groups"][offence_group][
                    "offence_subgroups"][offence_subgroup] = offence_subgroup_group["Count"].sum().item()
    # Move the "Unknown" borough to the end of the dictionary
    unknown_borough = processed_data["dates"][date_string]["boroughs"].pop("Unknown")
    processed_data["dates"][date_string]["boroughs"]["Unknown"] = unknown_borough

crime_data_json = json.dumps(processed_data, indent=4)
with open("src/london_crime_data.json", "w") as file:
    file.write(crime_data_json + "\n")
```
